```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
three = pd.read_csv('Documents/Capstone Project/Data/2003.csv')
four = pd.read_csv('Documents/Capstone Project/Data/2004.csv')
five = pd.read_csv('Documents/Capstone Project/Data/2005.csv')
six = pd.read_csv('Documents/Capstone Project/Data/2006.csv')
seven = pd.read_csv('Documents/Capstone Project/Data/2007.csv')
eight = pd.read_csv('Documents/Capstone Project/Data/2008.csv')
nine = pd.read_csv('Documents/Capstone Project/Data/2009.csv')
ten = pd.read_csv('Documents/Capstone Project/Data/2010.csv')
eleven = pd.read_csv('Documents/Capstone Project/Data/2011.csv')
twelve = pd.read_csv('Documents/Capstone Project/Data/2012.csv')
thirteen = pd.read_csv('Documents/Capstone Project/Data/2013.csv')
fourteen = pd.read_csv('Documents/Capstone Project/Data/2014.csv')
fifteen = pd.read_csv('Documents/Capstone Project/Data/2015.csv')
sixteen = pd.read_csv('Documents/Capstone Project/Data/2016.csv')
```

```python
frames = [three, four, five, six, seven, eight, nine, ten, eleven, twelve, thirteen, fourteen, fifteen, sixteen]
result = pd.concat(frames)
```

```python
aq = pd.DataFrame(result, columns = ['Date (Local)','Parameter Name','First Maximum Value','Duration Description'])
aq = aq.rename(index = str, columns = {'Date (Local)': 'date','Parameter Name':'parameter','First Maximum Value':'value','Duration Description':'duration'})
```

```python
aq = aq[aq.parameter != 'Acceptable PM2.5 AQI & Speciation Mass']
aq = aq[aq.duration != "1 HOUR"]
aq['date'] = pd.to_datetime(aq['date'])
aq.head()
```

```python
pm = aq[aq.parameter == 'PM2.5 - Local Conditions']
pm.describe()
```

```python
ozone = aq[aq.parameter == 'Ozone']
ozone.head()
```

```python
import seaborn as sns
```

```python
fig = plt.figure()
plt.plot(pm.date,pm.value)
plt.title('PM2.5 (24-hr avg) Over Time')
plt.ylabel('$\mu g/m^3$')
plt.plot(('2003-02-28', '2006-10-17'), (65, 65), 'r-')
plt.plot(('2006-10-17', '2016-12-31'), (35, 35), 'r-')
plt.plot(('2003-02-28', '2013-01-15'), (15,15), 'c-')
plt.plot(('2013-01-15', '2016-12-31'), (12,12), 'c-')
plt.show()
fig.savefig('PM25_2003_2016.png')
```

```python
fig = plt.figure()
plt.plot(ozone.date,ozone.value)
plt.title('$O_3$ (8-hr avg) Over Time')
plt.ylabel('ppm')
plt.plot(('2003-02-28', '2008-03-27'), (0.08, 0.08), 'r-')
plt.plot(('2008-03-27', '2015-10-26'), (0.075, 0.075), 'r-')
plt.plot(('2015-10-26', '2016-12-31'), (0.07, 0.07), 'r-')
plt.show()
fig.savefig('O3_2003_2016.png')
```

```python
frames = [three, four, five, six, seven, eight, nine, ten, eleven, twelve, thirteen, fourteen, fifteen, sixteen]
result = pd.concat(frames)
```

```python
south = pd.DataFrame(result, columns = ['Date (Local)','Parameter Name','First Maximum Value'])
south = south.rename(index = str, columns = {'Date (Local)': 'date','Parameter Name':'parameter','First Maximum Value':'value'})
south.head()
```

```python
south = south[south.parameter == 'PM2.5 - Local Conditions']
south['date'] = pd.to_datetime(south['date'])
south.describe()
```

```python
fig = plt.figure()
sns.distplot(pm.value, bins=50, color='r', label='North')
sns.distplot(south.value, bins=50, color='b', label='South')
plt.title('Distribution of PM2.5 Values')
plt.xlabel('$\mu g/m^3$')
plt.ylabel('Frequency')
plt.legend()
plt.show()
fig.savefig('PM_North_South_Hist.png')
```

```python
fig = plt.figure()
plt.hist(pm.value, bins=50, color='r', alpha=0.5, label = 'North')
plt.hist(south.value, bins=50, color='b', alpha=0.5, label = 'South')
plt.title('Distribution of PM2.5 Values')
plt.xlabel('$\mu g/m^3$')
plt.ylabel('Frequency')
plt.legend()
plt.xlim([35,80])
plt.ylim([0,5])
plt.show()
fig.savefig('PM_North_South_Worst.png')
```

```python
weather = pd.read_excel('Documents/Capstone Project/columbus_weather_cleaned2.xlsx')
weather.head()
```

```python
past_year_aq = pd.read_csv('Documents/Capstone Project/aq.csv')
past_year_aq['Date'] = pd.to_datetime(past_year_aq['Date'])
past_year_aq = pd.DataFrame(past_year_aq, columns = ['Date','o3','pm25'])
#past_year_aq = past_year_aq.reindex(past_year_aq['Date'])
past_year_aq.head()
```

```python
merged = past_year_aq.join(weather, on=past_year_aq['Date'], how='left')
merged.head()
```

```python
merged.describe()
```

```python
fig = plt.figure()
plt.plot(merged.o3)
plt.title('Hourly $O_3$ Concentration, March 2016 to March 2017')
plt.ylabel('ppm')
plt.show()
fig.savefig('ozone_past_year.png')
```

```python
fig = plt.figure()
plt.plot(merged.pm25)
plt.title('Hourly PM2.5 Concentration, March 2016 to March 2017')
plt.ylabel('$\mu g/m^3$')
plt.show()
fig.savefig('pm25_past_year.png')
```

```python
fig = plt.figure()
plt.plot(merged.TEMP)
plt.title('Hourly Temperature, March 2016 to March 2017')
plt.ylabel('Temp (Deg F)')
plt.show()
fig.savefig('temp_past_year.png')
```

```python
plt.plot(merged.SPD)
plt.show()
```
